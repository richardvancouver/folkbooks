## Low Temperature Measurements

For everything below here T ~ 7mK, at least at the mixing chamber. The sample temperature not well known, yet. 

```python
# SIN wire
# AuS3_LR_4770 (chip name - which corner - sole's RT resistance value)

Vac = 10e-6
condquant = 3.75e-5
amplification = 1e-6
multiplier = amplification/Vac/condquant

w = Waves('dat23g1x.ibw', y_multiplier = multiplier)

fig, ax = plt.subplots(1,1)
ax.plot(w.x/1000.0, w.y)
ax.set_xlabel('$V_{BG}\ (V)$')
ax.set_ylabel(r'$G\ (e^2/h)$')
```

### Switching to current biased measurement

Iac = 1nA

```python
# SIS wire Au3S_LL_2132

Iac = 1e-9
amplification = 1e-3
multiplier = amplification/Iac

w0 = Waves('dat55g1x.ibw', y_multiplier = multiplier)
w1 = Waves('dat56g1x.ibw', y_multiplier = multiplier)
w2 = Waves('dat57g1x.ibw', y_multiplier = multiplier)
u = Waves('dat58g2d1x.ibw', y_multiplier = multiplier)

fig, ax = plt.subplots(1,2, figsize = (14,6))
ax[0].plot(w0.x/1000.0, w0.y)
ax[0].plot(w1.x/1000.0, w1.y)
ax[0].plot(w2.x/1000.0, w2.y)
# ax[0].set_ylim(9, 11)
ax[0].set_xlabel('$V_{bias}\ (mV)$')
ax[0].set_ylabel('$R\ (k\Omega)$')

extent = [u.extent[0]/1000.0, u.extent[1]/1000.0, u.extent[2]/10.0, u.extent[3]/10.0]
im = ax[1].imshow(u.y, extent = extent, aspect = 'auto',
                  vmin = 5500.0, vmax = 7000.0, 
                  interpolation='nearest')
ax[1].set_xlabel('$V_{BG}\ (V)$')
ax[1].set_ylabel('$I_{bias}\ (nA)$')

cb = plt.colorbar(im)
cb.set_label('$k\Omega$')
fig.tight_layout()
```
