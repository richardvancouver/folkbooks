```python
def find_nearest(vals, data):
    inds = np.zeros(len(vals), dtype=np.int)
    for i,val in enumerate(vals):
        inds[i] = np.argmin(np.abs(data-val))
    return inds

w = Waves('dat109g2d1x.ibw', x_multiplier=(1e-3, 1e-3), y_multiplier=1e-7/1e-5/3.874e-5, transpose=False)

fig, ax = plt.subplots(1,1)

x = w.x[:,0,1]
gates = [-5.0, -4.9, -4.6, -4.2]
igates = find_nearest(gates, w.x[0,:,0])
for i, ig in enumerate(igates):
    ax.plot(x, w.y[:,i], lw=2, label = '{0} V'.format(gates[i]))

ax.legend(loc='lower left')
ax.set_xlim(-0.6, 0.6)
ax.set_ylim(0,None)
ax.grid()
ax.set_axisbelow(True)
ax.set_ylabel('G ($e^2 / h$)')
ax.set_xlabel('$V_{bias}$ (mV)')
```
