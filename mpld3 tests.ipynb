```python
%matplotlib inline
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import mpld3
```

```python
N = 25
imx = np.linspace(0,1,N)
imy = np.linspace(0,1,N)
IMX, IMY = np.meshgrid(imx, imy)
data = (np.cos(2*(IMX**2 + IMY**2)*np.pi) + np.sin(2*(IMX**2 + IMY**2)*np.pi))
extent = [imx.min(), imx.max(), imy.min(), imy.max()]

fig, ax = plt.subplots()
im = ax.imshow(data, cmap=plt.cm.viridis, interpolation='nearest', 
               extent=extent) 
# needs origin='lower' flag for matplotlib not for mpld3

ax.set_xlabel('x label (unit)')
ax.set_ylabel('y label (units)')
cb = plt.colorbar(im)
mpld3.display(fig)
```
