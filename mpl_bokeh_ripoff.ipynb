```python
%matplotlib inline
from cStringIO import StringIO
from igor.binarywave import Waves
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
mpl.rcParams['font.family'] = 'sans-serif'
mpl.rcParams['font.style'] = 'normal'
mpl.rcParams['font.size'] = 14
mpl.rcParams['font.sans-serif'] = 'arial'
```

```python
w = Waves('test_data/dat17g1x.ibw')

fig, ax = plt.subplots(1,1, figsize=(7,5), dpi=100)
ax.grid(which='major', color='#D3D3D3', linestyle='-')
ax.spines['top'].set_color('#D3D3D3')
ax.spines['right'].set_color('#D3D3D3')
ax.set_axisbelow(True)

ax.tick_params(direction='out')
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')

ax.plot(w.x, w.y, linewidth=2, color='#B22222')

ax.set_title('dat1234g1x', fontsize = 'x-large', color='#525252')
ax.set_xlabel('x label (units)', color='#525252')
ax.set_ylabel('y label (units)', color='#525252')

s = StringIO()
fig.savefig(s, dpi=100, format='pdf')
with open('test_pdf_2.pdf', 'w+') as fo:
    fo.write(s.getvalue())
```
